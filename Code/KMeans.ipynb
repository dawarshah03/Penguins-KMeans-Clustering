```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
df= pd.read_csv('penguins.csv')
df = df.dropna()
df.head()
```

```python
plt.scatter(df['culmen_length_mm'], df['culmen_depth_mm'])
```

```python
scaler = MinMaxScaler()
df[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']] = scaler.fit_transform(df[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']])
df.head()
```

```python
km=KMeans(n_clusters=3)
km
```

```python
x = df[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']]
y_predicted=km.fit_predict(x)
y_predicted
```

```python
df['cluster']=y_predicted
df.head(20)
```

```python
df0 = df[df.cluster == 0]
df1 = df[df.cluster == 1]
df2 = df[df.cluster == 2]

plt.scatter(df0['culmen_length_mm'], df0['culmen_depth_mm'], color='green', label='Cluster 0')
plt.scatter(df1['culmen_length_mm'], df1['culmen_depth_mm'], color='red', label='Cluster 1')
plt.scatter(df2['culmen_length_mm'], df2['culmen_depth_mm'], color='black', label='Cluster 2')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], color='blue', marker='x', label='Centroids')

plt.xlabel('Culmen Length (mm)')
plt.ylabel('Culmen Depth (mm)')
plt.legend()
plt.show()
```

```python
km.cluster_centers_
```

```python
k_rng=range(1,10)
sse=[]
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(df[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of Squared Error')
plt.plot(k_rng,sse)
```
